```python
from collections import Counter
from konlpy.tag import Okt
from konlpy.utils import pprint
import pandas as pd
```

```python
df = pd.read_csv('Senti/data_8074.csv')
```

```python
from wordcloud import WordCloud
```

```python
okt = Okt()
df['tokenized'] = df['icls_cnts'].apply(okt.nouns)
```

```python
stopwords = ['도','는','다','의','가','이','은','한','에','하','고','을','를','인','듯','과','와','네','들','듯','지','임','게','만','게임','겜','되','음','면', '경찰', '지난', '라며', '대해','경우','최근','모두','지난해','지난달','해당','일부','오전','다시','한편','다른','때문','대한','이번','위해']
```

```python
df['tokenized'] = df['tokenized'].apply(lambda x: [item for item in x if item not in stopwords and len(item) >1])
```

```python
import numpy as np

news = np.hstack(df['tokenized'].values)
```

```python
new_counts = Counter(news)
```

```python
temp_data = ' '.join(news)
```

```python
import matplotlib.pyplot as plt
```

```python
import os
#font = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
wc = WordCloud(width = 1600, height = 800, font_path = 'NanumBarunGothic').generate(temp_data)
plt.figure(figsize = (15,15))
plt.imshow(wc, interpolation = 'bilinear')
```
